```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,4,2,8,5]
y1 = [3,4,6,7,8]
plt.figure(figsize=(10,3), dpi=100)
plt.plot(x[:4],y[:4],'y*-.', label="USA")
plt.plot(x[3:],y[3:],'r*-.', label="USA")
plt.plot(x,y1, color="blue", marker=".", markeredgecolor="black", linestyle="-.", label="India")
plt.xlabel("Gas")
plt.ylabel("Price")
plt.title("Graph")
plt.legend()
plt.show()
plt.savefig("graph.png")
```

```python
label = ["A","B","C"]
values = [10,15,5]
bars = plt.bar(label,values, color = "red")

bars[0].set_facecolor('blue')
bars[0].set_hatch('*')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Bar Graph" , fontdict={'fontweight':'bold', 'fontname':'DejaVu Sans'})
plt.show()
```

```python
gas = pd.read_csv("gas_prices.csv")
print(gas)
```

```python
plt.plot(gas.Year,gas.Australia,'r.-', label="Australia")
plt.plot(gas.Year,gas.USA,'b.-', label="USA")
plt.plot(gas.Year,gas.Canada,'g.-', label="Canada")
plt.plot(gas.Year,gas.UK,'y.-', label="UK")
plt.plot(gas.Year,gas['South Korea'],'c.-', label="South Korea")
plt.plot(gas.Year,gas['Germany'],'m.-', label="Germany")
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("Gas Prices")
plt.legend()
plt.show()
```

```python

labels = ["A", "B", "C"]
values = [10, 15, 5]
plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Pie Chart Example')
plt.show()
```
